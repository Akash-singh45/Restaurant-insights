# üßπ Data Cleaning & Preprocessing



```python
df_cleaned = df.copy()
# making copy of original dataset.
```

```python
#Check for missing values
df_cleaned.isnull().sum()
```

```python
# remove duplicates
df_cleaned.duplicated().sum() # checking duplicates
df_cleaned.drop_duplicates(inplace=True)
```

```python
#stripping white spaces from column name.
df_cleaned.columns = df_cleaned.columns.str.strip()

for col in df_cleaned.select_dtypes(include='object').columns:
    df_cleaned[col] = df_cleaned[col].astype(str).str.strip()
```

```python
df_cleaned.head()
```

## üßπ Data Cleaning & Preprocessing

Steps Performed:
- Removed duplicates
- Cleaned column names and string values
- Checked for missing data
-removed whitespaces from column name


# **Task 1**

```python

# Determine the top three most common cuisines in the dataset.
top_cuisines = df_cleaned['Cuisines'].value_counts().head(3)
print("Top 3 most common cuisines:")
print(top_cuisines)
```

```python
#Calculate the percentage of restaurants that serve each of them.
total_restaurants = len(df_cleaned)
percentage_per_cuisine = round((top_cuisines / total_restaurants) * 100,2)
print("\nPercentage of restaurants that serve each of the top 3 cuisines:")
print(percentage_per_cuisine)
```

# Task 2

```python
#Identify the city with the highest number of restaurants in the dataset.
city_with_most_restaurants = df_cleaned['City'].value_counts().head(1)
print("City with the highest number of restaurants:", city_with_most_restaurants)
```

```python
# Calculate the average rating for restaurants in each city.
average_rating_per_city = df_cleaned.groupby('City')['Aggregate rating'].mean()
print("\nAverage rating for restaurants in each city:")
print(average_rating_per_city)


```

```python
# Determine the city with the highest average rating.
city_with_highest_Average_rating = average_rating_per_city.idxmax()
print("\nCity with the highest average rating:", city_with_highest_Average_rating)
```

# Task 3: Price Range **Distribution**

```python
#Visualize the distribution of price ranges using a histogram or bar chart.
```

```python
# Set a style
sns.set(style="whitegrid")

# Plot a bar chart of price range distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df_cleaned, x='Price range', palette='viridis')

# Set labels and title
plt.title('Distribution of Restaurant Price Ranges')
plt.xlabel('Price Range (1 = Low, 4 = High)')
plt.ylabel('Number of Restaurants')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Calculate the percentage of restaurants in each price range category.

price_range_counts = df_cleaned['Price range'].value_counts().sort_index()

# Total number of restaurants
total_restaurants = len(df_cleaned)

# Calculate percentage
price_range_percentage = (price_range_counts / total_restaurants) * 100

# Display results
print("Percentage of restaurants in each price range:")
print(price_range_percentage.round(2))  # rounding to 2 decimal places
```

# **Task 4: Online Delivery**

```python
#Determine the percentage of restaurants that offer online delivery.
online_delivery_counts = df_cleaned['Has Online delivery'].value_counts()
total_restaurants = len(df_cleaned)

# Calculate percentage
online_delivery_percentage = (online_delivery_counts / total_restaurants) * 100

# Display results
print("Percentage of restaurants that offer online delivery:")
print(online_delivery_percentage.round(2))  # rounding to 2 decimal places
```

```python
#Compare the average ratings of restaurants with and without online delivery.
rating_online_delivery = df_cleaned.groupby('Has Online delivery')['Aggregate rating'].mean()

# Display results
print("Average ratings of restaurants with and without online delivery:")
print(rating_online_delivery.round(2))  # rounding to 2 decimal places
```
